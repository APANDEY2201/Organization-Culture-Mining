```python
df_truncated = df[truncated_df_list]
df_truncated
```

```python
list_indx_coeff=[0,4,8,12,16,20,24]
list_indx_p_stars=[1,5,9,13,17,21,25]
df_coeff = df_truncated[df_truncated.index.isin(list_indx_coeff)]
df_p_stars = df_truncated[df_truncated.index.isin(list_indx_p_stars)]
df_coeff.reset_index(inplace=True)
df_p_stars.reset_index(inplace=True)
df_p_stars['Dependent Feature'] = df_coeff['Dependent Feature']
df_coeff.set_index('Dependent Feature', inplace=True)
df_p_stars.set_index('Dependent Feature', inplace=True)
```

```python
df_p_stars
```

```python
df_coeff_final = df_coeff[heatmap_features]
df_p_stars_final = df_p_stars[heatmap_features]
```

```python
df_coeff_final = df_coeff_final.astype(float)
```

```python
df_coeff_list = df_coeff_final.values.tolist()
df_p_stars_list = df_p_stars_final.values.tolist()
```

#### Formatting text to put inside heatmap cells

```python
for i in range(len(df_coeff_list)):
    for j in range(len(df_coeff_list[i])):
        df_p_stars_list[i][j] = str(df_coeff_list[i][j])+'\n'+str(df_p_stars_list[i][j])
```

#### Plotting Heatmap

```python
plt.rcParams.update({'font.size': 8})
fig = plt.figure(figsize=(55/2.54,15/2.54)) #dt2
ax = fig.add_axes([0.25,0.45,0.75,0.55]) # left, bot, width, height
cmap = sns.diverging_palette(10, 150, as_cmap=True)
sns.heatmap(df_coeff_final, cmap=cmap, linewidths=1, annot=df_p_stars_list, vmin=-1, vmax = 1 , cbar = False, square =False, annot_kws={"size": 8})
plt.savefig("test_heatmap_DS"+str(dataset)+"_all_features.jpeg")
```
