```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Manual Intervention
From the individual sector-wise csv files

make 1 csv of Dataset1 for all sectors

make 1 csv of Dataset2 for all sectors

# Input: Cumulative Topics Proportions csvs for Dataset1 and Dataset2

```
valuesDf1 = pd.read_csv("dataOCM/02_LDA/Cumulative_Topics_Proportions_SectorWise/RQ4_HeatMap_Dataset1_Acronyms.csv", index_col=[0])
valuesDf2 = pd.read_csv("dataOCM/02_LDA/Cumulative_Topics_Proportions_SectorWise/RQ4_HeatMap_Dataset2_Acronyms.csv", index_col=[0])
```

# Data for Heat Map

```
listA = [valuesDf1.values.tolist()[i] for i in range(0, len(valuesDf1.values.tolist()))]
listB = [valuesDf2.values.tolist()[i] for i in range(0, len(valuesDf2.values.tolist()))]
```

# Plotting Heat Map from listA and listB

```
plt.rcParams.update({'font.size': 8})
fig, axs = plt.subplots(2,1,figsize=(18/2.54, 23.67/2.54))
# fig = plt.figure(figsize=(18/2.54,23.67/2.54)) #dt1 10.45
# fig = plt.figure(figsize=(9/2.54,11.83/2.54)) #dt2
# axs[0].rect = [0.25,0.22,0.60,0.54]
# axs[0] = fig.add_axes([0.31,0.22,0.60,0.54])
axs[0].text(20.7/2.54, 23.67/2.54, 'Dataset1', fontsize=8, rotation=90)
axs[0].xaxis.tick_top()
# cmap = sns.diverging_palette(150, 10, as_cmap=True)
# print(cmap)
x_axis_labels = valuesDf1.columns
y_axis_labels = valuesDf1.index
cmap = sns.light_palette("seagreen", as_cmap=True)
#cmap = sns.diverging_palette(0, 109, s=43, l=59, sep=1, n=6, center='light', as_cmap=True)
#cmap = sns.diverging_palette(150, 10, as_cmap=True) used by Anuraag
hmap = sns.heatmap(listA, cmap=cmap, linewidths=0, annot=listA, cbar = False, square =False, ax = axs[0], annot_kws={"size": 8}, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
#cbar = axs[0].collections[0].colorbar
#cbar.set_ticks([4, 5, 6, 7, 8, 9])
#cbar.set_ticklabels(['4%', '5%', '6%', '7%', '8%', '9%'])
hmap.set_yticklabels(y_axis_labels, rotation=0)
axs[0].axhline(y=0, color='black',linewidth=1)
axs[0].axhline(y=18, color='black',linewidth=2)
#for i in range(18):
    #axs[0].axvline(x=i, color='black',linewidth=0.5)
axs[0].axvline(x=0, color='black',linewidth=1)
axs[0].axvline(x=8, color='black',linewidth=2)
# axs[1].rect = [0.25,0.22,0.60,0.54]
# axs[1] = fig.add_axes([0.31,0.22,0.60,0.54])
axs[1].text(20.7/2.54, 23.67/2.54, 'Dataset2', fontsize=8, rotation=90)
axs[1].get_xaxis().set_visible(False)
# cmap = sns.diverging_palette(150, 10, as_cmap=True)
# print(cmap)
x_axis_labels = valuesDf2.columns
y_axis_labels = valuesDf2.index
cmap = sns.light_palette("seagreen", as_cmap=True)
hmap = sns.heatmap(listB, cmap=cmap, linewidths=0, annot=listB, cbar = False, square =False, ax = axs[1], annot_kws={"size": 8}, xticklabels=x_axis_labels, yticklabels=y_axis_labels)
#cbar = axs[1].collections[0].colorbar
#cbar.set_ticks([2, 6, 10, 14, 18, 22])
#cbar.set_ticklabels(['2%', '6%', '10%', '14%', '18%', '22%'])
hmap.set_yticklabels(y_axis_labels, rotation=0)
axs[1].axhline(y=0, color='black',linewidth=1)
axs[1].axhline(y=20, color='black',linewidth=2)
#for i in range(18):
    #axs[1].axvline(x=i, color='black',linewidth=0.5)
axs[1].axvline(x=0, color='black',linewidth=1)
axs[1].axvline(x=8, color='black',linewidth=2)
plt.savefig("dataOCM/02_LDA/Cumulative_Topics_Proportions_SectorWise/Exp_5_Cumulative_Topics_Proportions_Sectorwise.png")
plt.show()
```
